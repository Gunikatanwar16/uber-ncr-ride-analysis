```python
df.drop_duplicates(inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['day'] = df['Date'].dt.day_name()
df['month'] = df['Date'].dt.month
df['year'] = df['Date'].dt.year
df['weekday'] = df['Date'].dt.weekday
```

```python
sns.countplot(x='Booking Status', data=df)
plt.xticks(rotation=45)
plt.title("Booking Status Distribution")
plt.show()
```

```python
sns.kdeplot(df['Booking Value'], fill=True)
plt.title("Booking Value Density")
plt.show()
```

```python
df['Booking Value'].skew()
```

```python
sns.boxplot(y='Booking Value', data=df)
plt.title("Booking Value Outliers")
plt.show()
```

```python
sns.boxplot(x='Vehicle Type', y='Ride Distance', data=df)
plt.xticks(rotation=45)
plt.title("Distance by Vehicle Type")
plt.show()
```

```python
df['Vehicle Type'].value_counts().plot(kind='bar')
plt.title("Most Preferred Vehicle Type")
plt.show()
```

```python
sns.countplot(x='Payment Method', data=df)
plt.xticks(rotation=45)
plt.title("Payment Method Distribution")
plt.show()
```

```python
sns.scatterplot(x='Ride Distance', y='Booking Value', data=df)
plt.title("Distance vs Fare")
plt.show()
```

```python
sns.countplot(x='Vehicle Type', hue='Booking Status', data=df)
plt.xticks(rotation=45)
plt.title("Booking Status by Vehicle Type")
plt.show()
```

```python
sns.countplot(x='Cancelled Rides by Customer', data=df)
plt.title("Customer Cancellations")
plt.show()

sns.countplot(x='Cancelled Rides by Driver', data=df)
plt.title("Driver Cancellations")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Cancelled Rides by Customer', data=df)
plt.title("Customer Cancellations")
plt.show()
```

```python
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
